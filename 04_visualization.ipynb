```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
sns.mpl.style.use('ggplot')
```

```python
np.random.seed(20180604)
```

## Randomly Generated Time Series

```python
ts = pd.Series(
    np.random.randn(1000),
    index=pd.date_range('1/1/2000', periods=1000)
)
ts = ts.cumsum()
ts.plot()
```

## The Fair Dataset

```python
fair_df = sm.datasets.fair.load_pandas().data
```

```python
fair_df.head()
```

```python
# point: mean by default
# line: 95% CI by default
sns.pointplot(data=fair_df, x='rate_marriage', y='affairs')
```

```python
sns.pointplot(data=fair_df, x='religious', y='affairs')
```

```python
fair_r_df = pd.DataFrame(
    data=np.corrcoef(fair_df.T),
    columns=fair_df.columns,
    index=fair_df.columns
)
```

```python
sns.heatmap(fair_r_df, center=0, square=True)
```
